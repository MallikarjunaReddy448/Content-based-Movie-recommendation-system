```python
import ast

# Define genre renaming dictionary
genre_mapping = {
    "Science Fiction": "Sci-Fi",
        "Anime": "Animation",
        "Violent": "Violence",
    "Action": "Action",
    "Comedy": "Comedy",
    # Add other genres here
}

# Function to handle genre renaming and parsing
def rename_genres(genre_list):
    # Check if the genre_list is a string (i.e., needs to be evaluated into a list)
    if isinstance(genre_list, str):
        genre_list = ast.literal_eval(genre_list)  # Convert string representation of list to list
    # Rename genres based on the mapping
    standardized_list = [genre_mapping.get(genre, genre) for genre in genre_list]
    return standardized_list

# Apply the function to the Genre column
Movies_dataset["Genre"] = Movies_dataset["Genre"].apply(rename_genres)

# Now, let's check the result
print(Movies_dataset["Genre"].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(Movies_dataset_combained)
tfidf_matrix
```

```python
def get_movie_recommendations(movie_title,genres, top_n = 20):
    user_input = f"{movie_title} {genres}"
    user_vector = tfidf.transform([user_input])  # Vectorize user input

    # Compute cosine similarity with all movies
    cosine_sim = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # Get the top N recommendations
    top_indices = cosine_sim.argsort()[-top_n:][::-1]
    recommended_movies = Movies_dataset.iloc[top_indices]['Movie Name'].tolist()

    return recommended_movies
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
from fuzzywuzzy import process

# Recalculate the sigmoid kernel (precomputed similarity matrix)
sigmoid = sigmoid_kernel(tfidf_matrix_2, tfidf_matrix_2)

# Movie indices for easy lookup
movie_indices = pd.Series(Movies_dataset.index, index=Movies_dataset['Movie Name'])

# Function to find the closest title using fuzzy matching
def find_closest_title(input_title):
    matches = process.extract(input_title, Movies_dataset['Movie Name'], limit=1)
    if matches and matches[0][1] >= 80:  # Threshold for fuzzy matching
        return matches[0][0]
    return None

# Main recommendation function
def recommend_movie(movie_name, genres=None, use_fuzzy=False):
    recommendations = []
    fuzzy_matches = []

    # Step 1: Check if the movie exists in the dataset
    if movie_name in movie_indices:
        # Get index and calculate recommendations based on sigmoid similarity
        user_input_idx = movie_indices[movie_name]
        sigmoid_score = list(enumerate(sigmoid[user_input_idx]))
        sorted_sigmoid_score = sorted(sigmoid_score, key=lambda x: x[1], reverse=True)
        recommended_indices = [i[0] for i in sorted_sigmoid_score[1:11]]
        recommendations = Movies_dataset['Movie Name'].iloc[recommended_indices].tolist()
    elif use_fuzzy:
        # Step 2: Handle fuzzy matching if enabled and exact match not found
        closest_title = find_closest_title(movie_name)
        if closest_title:
            fuzzy_matches.append(closest_title)
            # Get index and calculate recommendations for the closest title
            user_input_idx = movie_indices[closest_title]
            sigmoid_score = list(enumerate(sigmoid[user_input_idx]))
            sorted_sigmoid_score = sorted(sigmoid_score, key=lambda x: x[1], reverse=True)
            recommended_indices = [i[0] for i in sorted_sigmoid_score[1:11]]
            recommendations = Movies_dataset['Movie Name'].iloc[recommended_indices].tolist()

    # Step 3: If no title match, recommend based on genres
    if not recommendations and genres:
        # Vectorize the genres and calculate similarity
        user_genres_vector = tfidf_2.transform([genres])
        genre_similarity = sigmoid_kernel(user_genres_vector, tfidf_matrix_2).flatten()
        genre_top_indices = genre_similarity.argsort()[-10:][::-1]
        recommendations = Movies_dataset['Movie Name'].iloc[genre_top_indices].tolist()

    return {
        'recommendations': recommendations,
        'fuzzy_matches': fuzzy_matches
    }
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_movies(movie_title, genres=None, top_n=20):
    """
    Recommend movies based on a given title and optional genres, using cosine similarity.
    
    Parameters:
    - movie_title (str): The title of the movie to base recommendations on.
    - genres (str, optional): Genres to prioritize in recommendations.
    - top_n (int): Number of top recommendations to return.
    
    Returns:
    - list: Recommended movie titles.
    """
    # Combine the movie title and genres (if provided) into a single input string
    if genres:
        user_input = f"{movie_title} {genres}"
    else:
        user_input = movie_title
    
    # Vectorize the user input
    user_vector = tfidf.transform([user_input])  # Ensure tfidf is the trained vectorizer

    # Compute cosine similarity with all movie vectors
    cosine_sim = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # If genres are provided, re-prioritize recommendations by genre match
    if genres:
        # Compute similarity of user genres with movie genres in the dataset
        genre_vector = tfidf.transform([genres])
        genre_cosine_sim = cosine_similarity(genre_vector, tfidf_matrix).flatten()
        
        # Combine scores (title match + genre match)
        combined_score = cosine_sim + 0.5 * genre_cosine_sim  # Weight genre similarity
    else:
        combined_score = cosine_sim

    # Get the top N indices with the highest combined similarity scores
    top_indices = combined_score.argsort()[-top_n:][::-1]

    # Fetch recommended movie titles
    recommended_movies = Movies_dataset.iloc[top_indices]['Movie Name'].tolist()

    return recommended_movies
```

```python
from fuzzywuzzy import process
from sklearn.metrics.pairwise import cosine_similarity

def recommend_movies(movie_title, genres=None, top_n=20):
    """
    Recommend movies based on a given title and optional genres, using cosine similarity.
    
    Parameters:
    - movie_title (str): The title of the movie to base recommendations on.
    - genres (str, optional): Genres to prioritize in recommendations.
    - top_n (int): Number of top recommendations to return.
    
    Returns:
    - list: Recommended movie titles.
    """
    # Check if movie title exists in the dataset
    if movie_title not in Movies_dataset['Movie Name'].values:
        print(f"Movie '{movie_title}' not found in the dataset. Trying fuzzy matching.")
        
        # Fuzzy matching to find the closest match from the movie dataset
        closest_match = process.extractOne(movie_title, Movies_dataset['Movie Name'])
        
        if closest_match:
            print(f"Did you mean '{closest_match[0]}'? Providing recommendations based on this movie.")
            movie_title = closest_match[0]  # Use the closest match found by fuzzy matching
            user_input = f"{movie_title} {genres}" if genres else movie_title  # Create user input after fuzzy match
        else:
            print("No similar movie found. Providing genre-based recommendations.")
            if not genres:
                return []  # No recommendations possible without genres
            user_input = genres  # Fallback to genre-based recommendations
    else:
        user_input = f"{movie_title} {genres}" if genres else movie_title

    # Vectorize the user input (this line will not raise an error now)
    user_vector = tfidf.transform([user_input])

    # Compute cosine similarity with all movie vectors
    cosine_sim = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # If genres are provided, re-prioritize recommendations by genre match
    if genres:
        genre_vector = tfidf.transform([genres])
        genre_cosine_sim = cosine_similarity(genre_vector, tfidf_matrix).flatten()

        # Combine cosine similarity scores without any normalization
        combined_score = cosine_sim + 0.5 * genre_cosine_sim  # Weight genre similarity
    else:
        combined_score = cosine_sim

    # Get the top N indices with the highest combined similarity scores
    top_indices = combined_score.argsort()[-top_n:][::-1]

    # Exclude the input movie from recommendations
    if movie_title in Movies_dataset['Movie Name'].values:
        input_index = Movies_dataset[Movies_dataset['Movie Name'] == movie_title].index[0]
        combined_score[input_index] = -1  # Set the score to a very low value

    # Fetch recommended movie titles with additional details
    recommended_movies = Movies_dataset.iloc[top_indices][['Movie Name', 'Genre', 'Movie Release Year']].to_dict('records')
    recommended_movies_df = pd.DataFrame(recommended_movies, index=range(1, len(recommended_movies) + 1))
    return recommended_movies_df
```
